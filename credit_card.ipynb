### **Credit Card Data Analysis Project**

### Introduction – Credit Card Data Analysis Project
In this project, I explore and analyze a real-world credit card customer dataset to uncover patterns in customer demographics and financial behavior. The dataset includes features such as gender, education, income, employment history, and property/car ownership, along with a labeled column indicating whether a person was accepted for credit or not.

The main objective is to understand the structure of the data, perform data cleaning, and apply exploratory data analysis (EDA) to gain insights. This forms the foundation for future applications like credit scoring models, fraud detection systems, and risk assessment tools.

By visualizing trends and relationships in the data, we aim to answer questions such as:

     What kind of customers own cars or property?

     How is income distributed across education levels?

     Are there patterns between employment and income?

     What features may affect credit approval decisions?

This project showcases key data science skills including data wrangling, merging datasets, handling missing values, and visualizing distributions — all essential for real-world analytics and machine learning projects.

```python
import pandas as pd

credit_card["label"]=credit_card["label"].replace({"Yes":1, "No" :0})
credit_card["GENDER"]=credit_card["GENDER"].replace({"M":1, "F" :0})
credit_card["Car_Owner"]=credit_card["Car_Owner"].replace({"Y" :1, "N" :0})
credit_card["Propert_Owner"]=credit_card["Propert_Owner"].replace({"Y" :1, "N" :0})
credit_card
```

```python
categories= ['EDUCATION', 'Type_Income', 'Marital_status','Type_Occupation','Housing_type']  # change as needed

for col in categories:
    print(col)
    print(credit_card[col].unique())
```

```python
categories=[#EDUCATION
['Secondary / secondary special','Higher education','Lower secondary',
 'Incomplete higher'],
#Type_Income
['Commercial associate' ,'Pensioner' ,'Working' ,'State servant'],
#Marital_status
['Married' ,'Single / not married', 'Civil marriage', 'Separated' ,'Widow'],
#Type_Occupation
['Core staff' ,'Cooking staff', 'Laborers', 'Sales staff' ,'Accountants',
 'Managers', 'Cleaning staff' ,'Drivers', 'Low-skill Laborers',
 'High skill tech staff', 'IT staff', 'Waiters/barmen staff',
 'Security staff', 'Medicine staff' ,'Private service staff' ,'HR staff',
 'Secretaries', 'Realty agents'],
#Housing_type
['House / apartment', 'Rented apartment', 'Municipal apartment',
 'With parents', 'Co-op apartment', 'Office apartment']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_encoder=OrdinalEncoder(categories=categories)
income_encoder=ordinal_encoder.fit_transform(credit_card[['EDUCATION','Type_Income','Marital_status','Type_Occupation','Housing_type']])
changed_data=pd.DataFrame(income_encoder,columns=['edu_code','income_code','Marital_code','occupation_code','housing_code'])                                       
changed_data
```

```python
code_data=pd.concat([credit_card,changed_data],axis=1)
code_data.dtypes
```

```python
code_data=code_data.astype({'Annual_income':'int','edu_code':'int','income_code':'int','Marital_code':'int','occupation_code':'int','housing_code':'int'})
```

```python
code_data.dtypes
```

```python
code_data.drop(columns=['Type_Income','EDUCATION','Marital_status','Housing_type','Birthday_count','Type_Occupation'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

**Train Tesat split for Classfication**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape,y_test.shape
```

Let us standartization technique by using **StandardSCaler()**

```python
#standardization technique
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#use the fit_transfrom method to fit the selected columns into the set, it will learn the parameters
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# both the X_Train and X_test should be converted into dataframe for easy reading
x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
# we will import the accuracy, precision, confusion matrix and classification report for each ML method
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, confusion_matrix, f1_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model=LogisticRegression()
logistic_model.fit(x_train_scaled, y_train)
```

**Predict the test data**

```python
y_prediction_log = logistic_model.predict(x_test_scaled)

accuracy_log = accuracy_score(y_test,y_prediction_log)
precision_score_log = precision_score(y_test,y_prediction_log,average='binary', zero_division=1)
confusion_matrix_log = confusion_matrix(y_test,y_prediction_log)
classification_report_log = classification_report(y_test,y_prediction_log)
recall_score_log = recall_score(y_test,y_prediction_log)
f1_score_log = f1_score(y_test,y_prediction_log)


print("Confusion matrix:\n",confusion_matrix_log,"\n")
print("Accuracy:",round(accuracy_log*100,2),"%\n")
print("Precision:",round(precision_score_log*100,2),"%\n")
print("Recall score:",round(recall_score_log*100,2),"%\n")
print("f1 score:",round(f1_score_log*100,2),"%\n")
print(classification_report_log)
```
